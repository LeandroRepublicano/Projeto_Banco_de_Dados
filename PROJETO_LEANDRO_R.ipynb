```python
CREATE TABLE projeto.eventos (
    IDEvento INT,
    CaminhoEvento VARCHAR(47),
    Federacao VARCHAR(14),
    Data DATE,
    PaisEvento VARCHAR(17),
    EstadoEvento VARCHAR(3),
    CidadeEvento VARCHAR(48),
    NomeEvento VARCHAR(156)
);

CREATE TABLE projeto.levantamento_peso (
    IDEvento INT,
    Nome VARCHAR(41),
    Sexo VARCHAR(9),
    Equipamento VARCHAR(12),
    Idade INT,
    Divisao VARCHAR(48),
    PesoCorporalKg NUMERIC,
    ClassePesoKg VARCHAR(5),
    MelhorAgachamentoKg NUMERIC,
    MelhorSupinoKg NUMERIC,
    MelhorLevantamentoTerraKg NUMERIC,
    TotalKg NUMERIC,
    Posicao VARCHAR(2)
);
```

```python
#1 Quais são os três países com maior número de eventos de levantamento de peso registrados na base de dados?

pd.read_sql_query("SELECT PaisEvento, COUNT(*) AS TotalCompeticoes FROM projeto.eventos GROUP BY PaisEvento ORDER BY TotalCompeticoes DESC LIMIT 3;", con = conn)
```

```python
#2 Como é a distribuição média de peso corporal entre os competidores nas diferentes classes de peso?

pd.read_sql_query("SELECT ClassePesoKg, AVG(PesoCorporalKg) AS MediaPesoCorporal FROM projeto.levantamento_peso GROUP BY ClassePesoKg ORDER BY MediaPesoCorporal;", con = conn)
```

```python
#3 Em que estados dos Estados Unidos ocorrem mais competições e quantas são?

pd.read_sql_query("SELECT EstadoEvento, COUNT(*) AS TotalCompeticoes FROM projeto.eventos WHERE PaisEvento = 'USA' GROUP BY EstadoEvento ORDER BY TotalCompeticoes DESC LIMIT 5;", con = conn)
```

```python
#4 Qual é o recorde de levantamento terra entre os competidores e de quem é?
pd.read_sql_query("SELECT Nome,pesocorporalkg, ClassePesoKg, divisao, MelhorLevantamentoTerraKg FROM projeto.levantamento_peso WHERE MelhorLevantamentoTerraKg = (SELECT MAX(MelhorLevantamentoTerraKg) FROM projeto.levantamento_peso);", con = conn)
```

```python
#5 Qual é a competição com o maior número de participantes?
pd.read_sql_query("SELECT e.IDEvento, e.NomeEvento, COUNT(l.IDEvento) AS TotalParticipantes FROM projeto.eventos e JOIN projeto.levantamento_peso l ON e.IDEvento = l.IDEvento GROUP BY 1,2 ORDER BY TotalParticipantes DESC LIMIT 1;", con = conn)
```

```python
#6 Quantas competições foram realizadas em cada estado dos Estados Unidos?
pd.read_sql_query("SELECT EstadoEvento, COUNT(*) AS TotalCompeticoes FROM projeto.eventos WHERE PaisEvento = 'USA' GROUP BY EstadoEvento ORDER BY TotalCompeticoes DESC;", con = conn)
```

```python
#7 Quem são os 5 atletas com maior total de levantamento na categoria Open Senior?
pd.read_sql_query("SELECT Nome, TotalKg FROM projeto.levantamento_peso WHERE Divisao = 'Crossfit' ORDER BY TotalKg DESC LIMIT 5;", con = conn)
```

```python
#8 Qual é a média de idade dos competidores para cada tipo de equipamento utilizado?
pd.read_sql_query("SELECT Equipamento, AVG(Idade) AS MediaIdade FROM projeto.levantamento_peso GROUP BY Equipamento;", con = conn)
```

```python
#9 Quais são os 5 atletas com maior número de competições participadas?
pd.read_sql_query("SELECT Nome, sexo, divisao, COUNT(DISTINCT IDEvento) AS TotalCompeticoes FROM projeto.levantamento_peso GROUP BY 1,2,3 ORDER BY TotalCompeticoes DESC LIMIT 5;", con = conn)
```

```python
#10 QuaL o maximo de peso total levantado por competição ?
pd.read_sql_query("SELECT e.NomeEvento, e.Data, Max(lp.TotalKg) AS Peso_total_levantado FROM projeto.eventos e LEFT JOIN projeto.levantamento_peso lp ON e.IDEvento = lp.IDEvento WHERE lp.TotalKg IS NOT NULL GROUP BY 1,2 ORDER BY Peso_total_levantado DESC LIMIT 10;", con = conn)
```
